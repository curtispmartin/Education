## What is Programming? 

Well, programming (or coding) is our way of communicating w/ a digital device, like a computer or a phone. Usually this means we're bossing it around b/c, while computers have plenty of processing power, they have no understanding of how to use that power (i.e. they have no intuition), so we must give them instructions. 

This is generally quite difficult because computers & humans speak different languages: [Machine code](https://en.wikipedia.org/wiki/Machine_code) versus (e.g.) English. You can think of machine code a little bit like [what Neo sees after he realizes he's the One](https://youtu.be/Vy7RaQUmOzE?t=198):

<img src="Figures/matrixcode.png" alt="Git & GitHub" width="750" />

Like I said, kinda... But machine code is strictly numeric in nature. It isn't meant for humans, it's meant to give precise & quick instructions to the [central processing unit (CPU)](https://en.wikipedia.org/wiki/Central_processing_unit). So what we need is some sort of way to translate machine code into something which is human-readable. 

You might be thinking: Why not English? Or French? Or Japanese? The primary reason is that what we normally think of as languages -- that is the constructs we use to communicate w/ one another -- are very inefficient when it comes to giving instructions. If I asked you to explain to me how to create a new TikTok (or w/e you kids are doin these days) in written form, it would probably take a lot longer than it would if you just showed me, right? Not only that, but I'm liable to get vastly different instructions from each of you. That isn't what we want... It's incredibly inefficient.\*

**This brings us to programming languages, which are a sort of in-between: They're developed with words & symbols we recognize and are structured in a way we recognize, but also in a way which is more efficient & can be easily translated into machine code for execution.** These translations are done by computer programs called [compilers](https://en.wikipedia.org/wiki/Compiler). So what does this all mean? Well, like traditional languages, programming languages have rules which define *words* and *grammar* -- **What we call syntax.** This syntax (which the compiler translates) includes definitions for tasks like addition, multiplication, file writing, picture generation, data extraction, matrix manipulation, etc., & it's what we need to learn in order to do data science. 

\*Caveat: Don't get me wrong, it isn't as if we can only successfully write code in a single way. But the logic of programming languages typically constrains us in such a way so that the [cardinality](https://en.wikipedia.org/wiki/Cardinality) of the set of options isn't infinite.

### Operations
Now, if you asked (told?) Mr. Robot to `do`, what do you think would happen next? He'd ask you: **Do what?** (or if he was a good Mr. Robot he'd ask you **how high?** but that's neither here nor there). For example, you can tell him to:
- Add 5 & 5: `5+5`
- Print off a string: `print('Hello World!')`
- Calculate the mean of a set of numbers: `np.mean([1,2,3,4,5])` $\leftarrow$ we'll get to the `np.` part of this later...
- Read a file: `open(file.txt, 'r')`
- Plot some data: `sns.scatterplot(x=x, y=y, data=data)` $\leftarrow$ same here w/ `sns.`

These are all examples of what we call `operations`, or single actions (I'll use the terms interchangeably) performed by the computer, & they form the basis of our language. 

### Variables
But whenever you tell Mr. Robot to `do` something, the inevitable question he'll ask is: **to what?** The answer, of course, is to some sort of `object`, which we can assign a `variable` to represent. In a language, we might call this the subject of our statement. Assigning a variable is itself an action which is often representing using the equals sign `=`. So if I want to set a variable `x` to the object `5`, which in this case is a scalar, I'd write:
```
x=5
```
But what's the point? If I want to add 5 & 5, wouldn't `5+5` be easier? The answer of course is yes for verrrrrry simple & menial tasks. But what if I want to solve an equation? Plot a function? Read some text? Analyze a dataset? Build a neural network? Doing any of these things (& more!) generally requires many transformations of an object before reaching the desired outcome; the concept of objects & variables enables such transformations. 

A real-life (eventually!) example for you: Say you want to buy a house, but don't know what you can afford. Why not ask Mr. Robot to calculate how much your monthly mortgage payment will be? Assuming a 30-year fixed interest contract. 
```
### cost of house in dollars
price = 400000 
downpayment = 40000

### set some finance parameters
APR = 0.060
monthlyAPR = APR / 12
length = 30
payments = length * 12

### calculate monthly principal & interest payment
monthlypayment = (price - downpayment) * ((monthlyAPR * (1 + monthlyAPR) ** payments)) / (((1 + monthlyAPR) ** payments) - 1) 
```
And that's not counting any property taxes or insurance costs! But I digress... the point here is sure, you could type out the `monthlypayment` equation for each `price` you think is doable, but what a hassle that'd be to do more than a few times. Not to mention the huge potential for human error associated w/ that method, as well as the fact that you might forget it by the time you come back to it in a month. So instead of changing the numbers for Mr. Robot to make the calculation every time, you give him an equation & then tell him to change the numbers for you. It seems very simple, but it's worth reiterating just how much better the latter is. 

Keep in mind that `objects` don't have to be numeric; they can be strings ("like a sentence") or equations (m\*x+b) or functions (f(x,y))... More on that when we get into Python-specific programming!

### Iterables
Now, variables can point to different types of `objects`, which is also a subject for another class, but it behooves me here to introduce one specific type: An `iterable`. Recall that in math (specifically in linear algebra), variables have dimensionality to them. 
- A `scalar` is just a number; it has zero dimension (it can be thought of as akin to a point on a graph).
- A `vector` (or an `array` or a `list` or a `set` etc.) is a group of elements which all belong together, giving it a single dimension (it can be thought of as akin to a line on a graph). 

There are higher-dimensional variables as well, but for now what's important is that a `vector` is made up of a set of `scalars`, for which we can perform multiple different types of `operations`. Say for example, I asked you to take the mean of the variable `x`, where:
`x=[1,2,3,4,5]`

What would you do? Probably you'd calculate `(1+2+3+4+5)/5=7.5`, right? But what if I asked you to multiply `x` by 5? Then what? If I had to guess, I'd say you'd calculate `[1*5,2*5,3*5,4*5,5*5]=[5,10,15,20,25]`. `x` in this case is what we call an `iterable`: It's a variable which we can step through to perform an operation on the various elements which comprise it. What, then, would happen if I asked you to multiply `x*x`? We'll also wait for our Python-specific lectures to tackle this, but know that it depends on the program & the data type.

### Syntax
Think of this as a language's grammar. You may have noticed by now that whenever I finish a statement in my pseudocode, I start a new line. This is a general form of grammar, which is to say, we tend to give each statement its own line. Or in the case of compound sentences, like conditionals, we give each component of a statement a separate line. As we'll see, in Python, the new line character `\n` (which is what in MS Word, for example, defines a new line), is used to define the end of a statement; other programs use the semi-colon `;` or triangle brackets `<>` etc. But either way, part of the grammar of all programming languages is some character which tells Mr. Robot that a statement is finished (i.e. the period `.` of many languages).
